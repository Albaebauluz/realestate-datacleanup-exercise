#### Ejercicio 18. Normaliza la variable de precios para cada población y traza los 4 histogramas en el mismo gráfico (debes usar el subconjunto obtenido en la pregunta 14) (★★★)

Para el método de normalización, puedes usar el que consideres adecuado, no hay una única respuesta correcta para esta pregunta. Imprime el gráfico y escribe en la celda de Markdown un breve análisis sobre el gráfico.

Pista: Puedes ayudarte revisando la demostración multihist de Matplotlib.

```python
# Me quedo con los datos de Fuenlabrada, Leganés, Getafe y Alcorcón:
cinturon_sur = ds[ds['level5'].isin(['Fuenlabrada', 'Leganés', 'Getafe', 'Alcorcón'])]

# Normalizo la variable por el método Min-Max:
cinturon_sur["Normalized Price"] = (cinturon_sur['price'] - cinturon_sur['price'].min()) / (cinturon_sur['price'].max() - cinturon_sur['price'].min())

# Gráfico
# 1) Creo un gráfico con cuatro subplots:
fig, axes = plt.subplots(2, 2, figsize=(15, 10))

# 2) Creo un histograma para cada municipio:
municipios = ['Fuenlabrada', 'Leganés', 'Getafe', 'Alcorcón']
colors = ['blue', 'blue', 'blue', 'blue']

for ax, municipio, color in zip(axes.flatten(), municipios, colors):
    subset = cinturon_sur[cinturon_sur['level5'] == municipio]
    ax.hist(subset['Normalized Price'], bins=75, alpha=0.75, edgecolor=color, color=color)
    ax.set_title(f'Histograma del precio normalizado de la vivienda en {municipio}.')
    ax.set_xlabel('Precio')
    ax.set_ylabel('Frecuencia')

# Ajusto la disposición
plt.tight_layout()


plt.show()
```

Se puede observar que, tras normalizar los precios, esta variable siguie una distribución Chi-cuadrado en todos los municipios (con media en torno a 0.15) a excepción del precio de la vivienda en Getafe, que sigue una distribución normal, con media en torno a 0.2.

#### Ejercicio 19. ¿Qué puedes decir sobre el precio por metro cuadrado (precio/m2) entre los municipios de 'Getafe' y 'Alcorcón'? Debes usar el subconjunto obtenido en la pregunta 14 (★★☆)
